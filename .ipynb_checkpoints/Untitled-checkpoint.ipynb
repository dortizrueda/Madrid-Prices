```python
# 📦 Cargar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# 📥 Cargar el dataset
df = pd.read_csv("Datos.csv")

# 👀 Vista general de los datos
df.head()
```

```python
# 🧹 Información general
df.info()

# 🧼 Comprobación de valores nulos
df.isnull().sum().sort_values(ascending=False)
```

```python
# Convertir 'ascensor' a tipo booleano (si es 'Sí' o 'No')
df['ascensor'] = df['ascensor'].map({'Sí': True, 'No': False})

# Llenar valores faltantes simples
df['habitaciones'] = df['habitaciones'].fillna(df['habitaciones'].median())
df['ascensor'] = df['ascensor'].fillna(False)
df['ascensor'] = df['ascensor'].infer_objects(copy=False)
df['tags'] = df['tags'].fillna("Sin información")
df['descripcion'] = df['descripcion'].fillna("Sin descripción")
# Rellenar valores faltantes en 'localizacion' y 'planta'
df['localizacion'] = df['localizacion'].fillna("Sin localización")
df['planta'] = df['planta'].fillna("Sin planta")
```

```python
# 🧹 Información general
df.info()

# 🧼 Comprobación de valores nulos
df.isnull().sum().sort_values(ascending=False)
```

```python
# 🔍 Estadísticas descriptivas para columnas numéricas
df.describe()
```

```python
# 💰 Distribución de precios
sns.histplot(df['PrecioActual'], kde=True, bins=50)
plt.title("Distribución de precios de viviendas en Madrid")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# 📏 Superficie vs Precio
df_filtrado = df[df['metros'] < 500]

sns.scatterplot(data=df_filtrado, x="metros", y="PrecioActual", hue="zona", alpha=0.6)
plt.title("Relación entre superficie y precio")
plt.xlabel("Superficie (m²)")
plt.ylabel("Precio (€)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
top_zonas = df['zona'].value_counts().head(15)

sns.barplot(y=top_zonas.index, x=top_zonas.values)
plt.title("Top 15 zonas con más viviendas anunciadas")
plt.xlabel("Número de viviendas")
plt.ylabel("Zona")
plt.show()
```

```python
# 🔗 Correlaciones numéricas
corr = df.select_dtypes(include=[np.number]).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Mapa de calor de correlaciones")
plt.show()
```

```python
df_evo = pd.DataFrame({
    'PrecioActual': [150000, 175000, 200000, 250000, 280000, 220000, 300000, 350000, 450000, 550000],
    'fecha': ['2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01',
              '2021-05-01', '2022-03-01', '2022-07-01', '2023-01-01']
})

# Asegurarse de que la columna de fecha sea del tipo datetime
df_evo['fecha'] = pd.to_datetime(df_evo['fecha'])  # Cambié df por df_evo aquí

# Extraer solo el año de la fecha
df_evo['año'] = df_evo['fecha'].dt.year  # Cambié df por df_evo aquí

# Agrupar por año y calcular el precio promedio
df_evolucion = df_evo.groupby('año')['PrecioActual'].mean().reset_index()  # Cambié df por df_evo aquí

# Graficar la evolución de los precios
plt.figure(figsize=(10, 6))
plt.plot(df_evolucion['año'], df_evolucion['PrecioActual'], marker='o', color='b', linestyle='-', linewidth=2, markersize=6)
plt.title('Evolución de los Precios a lo largo de los Años', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Precio Promedio (EUR)', fontsize=12)
plt.grid(True)
plt.xticks(df_evolucion['año'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
from geopy.geocoders import Nominatim

# Tu DataFrame con las zonas y precios
data = {
    'zona': ['arganzuela', 'barajas', 'barrio-de-salamanca', 'carabanchel', 'centro', 
             'chamartin', 'chamberi', 'ciudad-lineal', 'fuencarral', 'hortaleza', 
             'latina', 'moncloa', 'moratalaz', 'puente-de-vallecas', 'retiro', 
             'san-blas', 'tetuan', 'usera', 'vicalvaro', 'villa-de-vallecas', 'villaverde'],
    'PrecioActual': [484212, 555625, 1966802, 273038, 862305, 1767126, 1498336, 542408, 
                     892204, 1190755, 299185, 1620871, 296772, 191280, 1310154, 
                     401028, 577429, 252342, 356815, 298188, 198630]
}

df_geo = pd.DataFrame(data)

# Geocodificador
geolocator = Nominatim(user_agent="geopandas")

# Crear listas para latitud y longitud
latitudes = []
longitudes = []

# Obtener coordenadas para cada zona
for index, row in df_geo.iterrows():
    location = geolocator.geocode(row['zona'] + ", Madrid, España")  # Agregar Madrid para mejorar la búsqueda
    if location:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(None)  # Si no se encuentra la coordenada
        longitudes.append(None)

# Agregar las coordenadas al DataFrame
df_geo['Latitud'] = latitudes
df_geo['Longitud'] = longitudes

# Mostrar el DataFrame actualizado con coordenadas
print(df_geo)
```

```python
# Verifica los nombres de las columnas en ambos DataFrames
print(df.columns)
print(df_geo.columns)
# Hacer el merge de df (datos originales) con df_geo (latitud y longitud)
df_final = pd.merge(df, df_geo[['zona', 'Latitud', 'Longitud']], on='zona', how='left')
print(df_final.columns)
```
